# Tema3- Jocuri cu adversari rationali

Clasa _Harta_ initializeaza o harta folosind constructorul ce primeste ca parametru fisierul de intrare si contine informatiile despre harta: tabelul (labirintul), inaltimea, latimea acestuia, pozitiile fantomelor, pozitiile mancarii, pozitia lui pacman.

```python
FOOD_SCORE = 20
LOOSE = "Loose"
PACMAN = "Pacman"
GHOST = "Ghost"

class Harta:
    def __init__(self, fin):
        self.table = []
        with open(fin) as file:
            line = file.readline()
            self.height, self.width = [int(res.strip()) for res in line.split(" ")]
            self.table = []
            self.ghosts = []
            self.food = []
            self.pacman = None
            no_ghosts = 0
            for i in range(0, self.height):
                line = file.readline()
                table_line = []
                for j in range(0, self.width):
                    ch = line[j]
                    if ch == '_':
                        table_line.append(True)
                    elif ch == 'X':
                        table_line.append(False)
                    elif ch == 'M':
                        table_line.append(True)
                        self.food.append((i, j))
                    elif ch == 'F':
                        no_ghosts += 1
                        table_line.append(True)
                        self.ghosts.append((i, j))
                    elif ch == 'P':
                        table_line.append(True)
                        self.pacman = (i, j)
                self.table.append(table_line)
```

```python
from copy import deepcopy
#state
PLAYERS = 0
FOOD = 1
IND = 2
SCORE = 3
LIVES = 4
#action
NAME = 0
MOVE = 1

name = ['Pacman', 'Ghosts']

def print_state(state, harta):
    string = []
    for i in range(0, harta.height):
        for j in range(0, harta.width):
            ch = ''
            if(harta.table[i][j]):
                ch = '_'
            else:
                ch = 'X'
            string.append(ch)
        string.append('\n')
    pacpos = state[PLAYERS][0]
    (i, j) = pacpos
    string[i * (harta.width + 1) + j] = 'P'
    for player in state[PLAYERS][1:]:
        (i, j) = player
        string[i * (harta.width + 1) + j] = 'F'
    for food in state[FOOD]:
        (i, j) = food
        string[i * (harta.width + 1) + j] = 'M'
    string = ''.join(string) + str("UrmeazÄƒ: %d - %s" % (state[IND], name[state[IND]]))
    string += '\nPozitie Pacman: ' + str(pacpos) + ", Scor: " + str(state[SCORE])\
                    + ", Vieti: " + str(state[LIVES])
    string += '\nPozitii Fantome: ' + str(state[PLAYERS][1:]) + '\n'
    return string

def init_state(harta, starter = 0):
    players = [harta.pacman] + harta.ghosts
    return (tuple(players), tuple(harta.food), starter, 0, 3)
    
def print_action(actions):
    string = ""
    for action in actions:
        string += str(action[NAME] + " moved " + action[MOVE]) + "\n"
    return string[:-1]

def get_actions(state, harta, dict_states = None):
    crt_player = state[IND]
    if crt_player == 0:
        crt_player = [crt_player]
    else:
        crt_player = [i for i in range(1, len(state[PLAYERS]))]
    
    actions = [tuple()]
    for player in crt_player:
        (i, j) = state[PLAYERS][player] 
        name = GHOST
        if player == 0:
            name = PACMAN
        else:
            name += str(player)
        new_actions = []
        for action in actions:
            if(i > 0 and harta.table[i-1][j]):
                new_actions.append(action + ((name, 'N'),));
            if (i < harta.height-1 and harta.table[i+1][j]):
                new_actions.append(action + ((name, 'S'),));
            if (j > 0 and harta.table[i][j-1]):
                new_actions.append(action + ((name, 'E'),));
            if (j < harta.width - 1 and harta.table[i][j+1]):
                new_actions.append(action + ((name, 'V'),));
        actions = new_actions
        
    if dict_states:
        new_actions = []
        for action in actions:
            if apply_action(state, action) in dict_states:
                continue
            new_actions.append(action)
        actions = new_actions
    return actions

def apply_action(state, actions):
    crt_player = state[IND]
    score = state[SCORE]
    lives = state[LIVES]
    if crt_player == 0:
        move = actions[0][MOVE]
        food = state[FOOD]
        player = state[PLAYERS][0]
        if move == 'N':
            player = (player[0]-1, player[1])
        elif move == 'S':
            player = (player[0] +1, player[1])
        elif move == 'E':
            player = (player[0], player[1]-1)
        elif move == 'V':
            player = (player[0], player[1] + 1)
        new_food = tuple(filter(lambda fpos: fpos != player, food)) 
        if(len(new_food) < len(food)):
            score += FOOD_SCORE
        players = (player,) + tuple(state[PLAYERS][1:])
        for i in range(1, len(players)):
            if players[0] == players[i]:
                lives -=1
                break
        return (players, new_food, 1, score, lives)
    else:
        players = []
        eaten = False
        
        for ind in range(len(actions)):
            move = actions[ind][MOVE]
            player = state[PLAYERS][1 + ind]
            if move == 'N':
                player = (player[0]-1, player[1])
            elif move == 'S':
                player = (player[0] +1, player[1])
            elif move == 'E':
                player = (player[0], player[1]-1)
            elif move == 'V':
                player = (player[0], player[1] + 1)
            players.append(player)
            if player == state[PLAYERS][0] and not eaten:
                eaten = True
                lives -= 1
        players = (state[PLAYERS][0],) + tuple(players)
        return (players, tuple(state[FOOD]), 0, score, lives)
    
def isFinal(state):
    players = state[PLAYERS]
    if state[LIVES] == 0:
        return 2
    if len(state[FOOD]) == 0:
        return 1
    return False
```

## 2) Pacman si fantomele sunt agenti aleatori

```python
MAX_ROUNDS = 3 * 2000

from random import randrange

def randomPlay(file):
    harta = Harta(file)
    s0 = init_state(harta)
    state = s0
    history_actions = []
    winner = False
    for _ in range(0, MAX_ROUNDS):
        actions = get_actions(state, harta)
        action = actions[randrange(len(actions))]
        history_actions.append(action)
        state = apply_action(state, action)
        winner = isFinal(state)
        if winner:
            print(("Pacman has won!" if winner==1 else "Ghosts have won!"))
            break
    if not winner:
        print("Nobody won!")
    for action in history_actions:
        print(print_action(action))
    print(print_state(state, harta))
randomPlay("fin0.txt")
```

## 3) Minimax cu alpha-beta

```python
MAX_LEVEL = 16

LOOP = "LOOP"

def manhattan(pos1, pos2):
    return abs(pos1[0] - pos2[0]) + abs(pos1[1] - pos2[1])

def evaluate1(state):
    pacman = state[PLAYERS][0]
    #scorul curent facut de pacman
    crt_score = state[SCORE]
    winner = isFinal(state)
    if winner == 2:
        return -float("inf")
    elif winner == 1:
        return crt_score
    #cu cat cea mai apropiata mancare e mai departata, cu atat scade mai mult scorul
    dist_min_food = -1.5 * min([manhattan(f, pacman) for f in state[FOOD]])
    #cu cat distanta de la pacman la cea mai apropiata fantoma e mai mare, cu atat
    #scorul este mai putin negativ iar cu cat este mai aproape cu atat este mai negativ
    dist_ghosts_pacman = -2 * \
        (1.0 / (1.0 + min([manhattan(ghost, pacman) for ghost in state[PLAYERS][1:]])))
    #pacman trebuie sa minimizeze numarul de mancare ramasa
    nr_food_left = -4 * len(state[FOOD])
    return crt_score + dist_min_food + dist_ghosts_pacman + nr_food_left

def evaluate2(state):
    return state[SCORE]
    
def minimaxPlay(harta, evaluate, max_level, s0, verbose = True, last_action = False):
    if s0[IND] == 0:
        val = -float("inf")
    else:
        val = float("inf")
        
    dict_score = {}
    for i in range(2, max_level):
        if s0[IND] == 0:
            new_val = maxPlay(harta, s0, -float("inf"), float("inf"),\
                                   i, 1, evaluate, dict_score)
            val = max([val, new_val])
        else:
            new_val = minPlay(harta, s0, -float("inf"), float("inf"),\
                                   i, 1, evaluate, dict_score)
            val = min([val, new_val])
    actions = dict_score[val][0]
    state = dict_score[val][1]
    
    if last_action:
        if actions == []:
            print(print_state(state))
            return False
        return actions[0]
    
    if verbose:
        if val < 0:
            print("Ghosts have won!")
        elif val > 0:
            print("Pacman has won!")
        else:
            print("Draw!")
        print("Score: " + str(val))
        print("Actions: ")
        for action in actions:
            print(print_action(action))
        print(print_state(state, harta))
    return val
        
def maxPlay(harta, state, alpha, beta, max_level, level, evaluate,\
            dict_score = {}, actions = []):
    winner = isFinal(state)
    if winner:
        score = state[SCORE]
        if score in dict_score and len(dict_score[score][0]) <= len(actions):
            return score
        dict_score[score] = (actions, state)
        return score
    if level == max_level:
        eval_score = evaluate(state)
        if eval_score in dict_score and len(dict_score[eval_score][0]) <= len(actions):
            return eval_score
        dict_score[eval_score] = (actions, state)
        return eval_score
    
    for action in get_actions(state, harta):
        new_state = apply_action(state, action)
        new_alpha = minPlay(harta, new_state, \
            alpha, beta, max_level, level + 1, evaluate,\
                           dict_score, actions + [action])
        alpha = max([alpha, new_alpha])
        if alpha >= beta:
            return beta
        
    return alpha
    
def minPlay(harta, state, alpha, beta, max_level, level, evaluate,\
           dict_score = {}, actions = []):
    winner = isFinal(state)
    if winner:
        score = state[SCORE]
        if score in dict_score and len(dict_score[score][0]) <= len(actions):
            return score
        dict_score[score] = (actions, state)
        return score
    if level == max_level:
        eval_score = evaluate(state)
        if eval_score in dict_score and len(dict_score[eval_score][0]) <= len(actions):
            return eval_score
        dict_score[eval_score] = (actions, state)
        return eval_score
    
    for action in get_actions(state, harta):
        new_state = apply_action(state, action)
        new_beta = maxPlay(harta, new_state, \
            alpha, beta, max_level, level + 1, evaluate,\
                          dict_score, actions + [action])
        beta = min([beta, new_beta])
        if alpha >= beta:
            return alpha
        
    return beta
```

## 4) Monte Carlo Tree Search

```python
N = 'N'
Q = 'Q'
PARENT = 'parent'
ACTIONS = 'actions'
```

```python
from random import choice

# FuncÈ›ie ce Ã®ntoarce un nod nou,
# eventual copilul unui nod dat ca argument
def init_node(parent = None):
    return {N: 0, Q: 0, PARENT: parent, ACTIONS: {}}

from math import sqrt, log

CP = 1.0 / sqrt(2.0)

# FuncÈ›ie ce alege o acÈ›iune dintr-un nod
def select_action(node, state, c = CP):
    N_node = node[N]
    # TODO <2>
    # Se cautÄƒ acÈ›iunea a care maximizeazÄƒ expresia:
    # Q_a / N_a  +  c * sqrt(2 * log(N_node) / N_a)
    expr_res = -1
    a_res = -1
    for a in node[ACTIONS]:
        np = node[ACTIONS][a]
        qa = np[Q]
        na = np[N]
        expr = qa / na + c * sqrt(2 * log(N_node) / na)
        if expr > expr_res:
            expr_res = expr
            a_res = a
            
    if a_res == -1:
        return False
    return a_res

# Algoritmul MCTS (UCT)
#  state0 - starea pentru care trebuie aleasÄƒ o acÈ›iune
#  budget - numÄƒrul de iteraÈ›ii permis
#  tree - un arbore din explorÄƒrile anterioare
#  opponent_s_action - ultima acÈ›iune a adversarului

def simulateRandom(harta, state):
    if isFinal(state):
        return state
    actions = get_actions(state, harta)
    
    while actions:
        action = choice(actions)
        new_state = apply_action(state, action)
        res = simulateRandom(harta, new_state)
        if res:
            return res
        actions.remove(action)
    if actions == []:
        return False
    
def mcts(harta, state0, budget, simulateFct):
    tree = init_node()
    
    for x in range(budget):
        state = state0
        node = tree

        while not isFinal(state) and \
            len(get_actions(state, harta)) == len(node[ACTIONS]):
            a = select_action(node, state)
            state = apply_action(state, a)
            node = node[ACTIONS][a]
        
        if not isFinal(state):
            all_actions = get_actions(state, harta)
            for a in all_actions:
                if a not in node[ACTIONS]:
                    state = apply_action(state, a)
                    node[ACTIONS][a] = init_node(node)
                    node = node[ACTIONS][a]
                    break
                
        if simulateFct == simulateRandom:
            state = simulateRandom(harta, state)
            winner = (isFinal(state)-1) if state else False
                
        elif simulateFct == minimaxPlay:
            reward = minimaxPlay(harta, evaluate1, 6, state, False)
            if reward > 0:
                winner = state0[IND]
            elif reward < 0:
                winner = 1 - state0[IND]
            else:
                winner = False
        else:
            reward = simulateFct(state)
            if reward > 0:
                winner = state0[IND]
            elif reward < 0:
                winner = 1 - state0[IND]
            else:
                winner = False
        
        if winner == state0[IND]:
            reward = 1
        elif winner == 1 - state0[IND]:
            reward = 0.0
        else:
            reward = 0.5    
                
        while node != None:
            node[N] += 1
            node[Q] += reward
            node = node[PARENT]

    final_action = select_action(tree, state0, 0.0)
    return final_action
```

```python
def randomAction(harta, state):
    actions = get_actions(state, harta)
    return choice(actions)

names = ["Monte-Carlo", "Monte-Carlo+Minimax", "Minimax", "Random"]
players = [mcts, mcts, minimaxPlay, randomAction]

def play_games(games_no, file, player1, player2,\
               max_rounds, euristica, budget):
    harta = Harta(file)
    
    algs = [player1, player2]
    if player1 == player2:
        name = [names[player1] + '0', names[player2] + '1']
    else:
        name = [names[player1], names[player2]]
    score = {p : 0 for p in name}
    
    print("%s vs %s \n" % (name[0], name[1]))
    for i in range(games_no):
        print("Jocul " + str(i+1))
        
        # Se desfÄƒÈ™oarÄƒ jocul
        state = init_state(harta)
        last_action = None        
        
        for _ in range(max_rounds):
            params = [[harta, state, budget, simulateRandom], \
              [harta, state, budget, minimaxPlay],\
              [harta, evaluate1, budget, state, False, True],\
              [harta, state]]
            p = state[IND]
            action = players[algs[p]](*params[algs[p]])
            if action == False:
                last_action = False
                break
            state = apply_action(state, action)
            last_action = action
            if isFinal(state):
                break
        
        print("Ultima stare: ")
        print(print_state(state, harta))
        
        # Cine a cÃ¢È™tigat?
        if last_action:
            eval_score = euristica(state)
            if eval_score > 0:
                winner = name[0]
                score[winner] += 1
                print("A cÃ¢È™tigat %s, scorul fiind %s\n" % (winner, eval_score))
            elif eval_score < 0:
                winner = name[1]
                score[winner] += 1
                print("A cÃ¢È™tigat %s, scorul fiind %s\n" % (winner, eval_score))
            else:
                print("Remiza\n")
        else:
            print("Remiza, se va repeta starea")

    # AfiÈ™eazÄƒ scorul final
    print("Scor final: %s.\n\n" % (str(score)))
```
